```python
import pandas as pd
import random
import math

data = pd.read_csv("./trainSignal.csv")
test_data = pd.read_csv("./testSignal.csv").rename(columns={'x': 'TARGET'})

example_ids = test_data['example_id'].values
x_test = test_data['TARGET'].values

x = data['x'].values
y = data['TARGET'].values
```

```python
def y_predict(a, b, c, d, x):
    return a * math.cos(b * x + c) + d

lr = 0.05
n = len(x)
batch = 100
epochs = 18000
rsl = []
a = random.random()
b = random.random()
c = random.random()
d = 0

for i in range(epochs):
    a_gradiente = 0
    b_gradiente = 0
    c_gradiente = 0
    d_gradiente = 0
    e = 0

    for m in range(batch):
        ix = int(random.uniform(0, n))
        error = y[ix] - y_predict(a, b, c, d, x[ix])
        e += error ** 2
        
        a_gradiente += -2 * math.cos(b * x[ix] + c) * error
        b_gradiente += -2 * x[ix] * math.sin(b * x[ix] + c) * error
        c_gradiente += -2 * math.sin(b * x[ix] + c) * error
        d_gradiente += -2 * error

    a -= lr * a_gradiente / batch
    b -= lr * b_gradiente / batch
    c -= lr * c_gradiente / batch
    d -= lr * d_gradiente / batch

    e /= batch
    rsl.append([a, b, c, d, e])
    print(f"error:{e} period:{b} amplitude:{a} constant: {c}")
```

```python
import matplotlib.pyplot as plt

def predict_signal(a, b, c, d, x_values):
    predicted_signal = [y_predict(a, b, c, d, xi) for xi in x_values]
    return np.array(predicted_signal)

x_values = np.arange(0, 20 * np.pi, 0.01)

epochs_to_visualize = [100, 500, 1000] + list(range(2000, 18000, 1000))

num_rows = len(epochs_to_visualize) // 3 + (len(epochs_to_visualize) % 3 > 0)
fig, axes = plt.subplots(num_rows, 3, figsize=(15, 5*num_rows))
axes = axes.flatten()

for i, epoch_to_visualize in enumerate(epochs_to_visualize):
    a, b, c, _, _ = rsl[epoch_to_visualize]
    
    predicted_signal = predict_signal(a, b, c, 0, x_values)
    
    real_x_values = data['x']
    real_y_values = data['TARGET']
    
    axes[i].plot(x_values, predicted_signal, color='blue', label='Predicted Signal')
    axes[i].plot(real_x_values, real_y_values, color='red', label='Real Data')
    axes[i].set_xlabel('x')
    axes[i].set_ylabel('TARGET')
    axes[i].set_title(f'Epoch = {epoch_to_visualize}')
    axes[i].legend()

for i in range(len(epochs_to_visualize), len(axes)):
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```
