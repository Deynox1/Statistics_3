```python
def shapiroTest(data):
    stat, p = stats.shapiro(data)
    alpha = 0.05

    if p > alpha:
        print("Los datos parecen seguir una distribucin normal (no se rechaza la hiptesis nula)")
    else:
        print("Los datos no siguen una distribucin normal (se rechaza la hiptesis nula)")
```

```python
shapiroTest(data['MSRP'])
shapiroTest(data['Invoice'])
shapiroTest(data['EngineSize'])
shapiroTest(data['Cylinders'])
shapiroTest(data['Horsepower'])
shapiroTest(data['MPG_City'])
shapiroTest(data['MPG_Highway'])
shapiroTest(data['Weight'])
shapiroTest(data['Wheelbase'])
shapiroTest(data['Length'])
```

```python
plt.figure(figsize=(12, 6))
plt.boxplot([data[data['Type'] == t]['MPG_City'] for t in data['Type'].unique()],
            labels=data['Type'].unique(), vert=False)
plt.title('Boxplot de MPG en Ciudad por Tipo de Vehculo')
plt.xlabel('MPG en Ciudad')
plt.ylabel('Tipo de Vehculo')
plt.show()
```

El grfico de caja muestra que la mayora de los datos estn cerca del medio en la mayora de los autos, pero en los Sedanes hay algunos valores extremos que estn muy lejos de la mayora, incluso fuera del rango normal. En cambio, en los autos hbridos, la mayora de los datos estn en el extremo ms bajo.

```python
plt.figure(figsize=(14, 6))
plt.boxplot([data[data['Make'] == make]['MPG_City'] for make in data['Make'].unique()],
            labels=data['Make'].unique(), vert=False)
plt.title('Boxplot de MPG en Ciudad por Fabricante (Make)')
plt.xlabel('MPG en Ciudad')
plt.ylabel('Fabricante (Make)')
plt.show()
```

Las grficas varan mucho entre las diferentes marcas de automviles. En algunos casos, como Honda, hay solo un valor atpico, pero en otros, como Kia, la mayora de los datos se encuentran en el extremo inferior. Los grficos de caja muestran que los valores de las millas por galn (MPG_City) son similares en la mayora de las marcas, pero algunas marcas tienen valores muy diferentes en este aspecto.

```python
plt.figure(figsize=(14, 6))
plt.boxplot([data[data['Origin'] == origin]['MPG_City'] for origin in data['Origin'].unique()],
            labels=data['Origin'].unique(), vert=False)
plt.title('Boxplot de MPG en Ciudad por Origen (Origin)')
plt.xlabel('MPG en Ciudad')
plt.ylabel('Origen (Origin)')
plt.show()
```

Los valores promedio de MPG_City son muy similares para los diferentes lugares de origen, lo que indica que la eficiencia de las millas por galn de los vehculos no est relacionada con su lugar de construccin. Sin embargo, en todos los casos, se observan varios valores atpicos que son excepciones a esta tendencia.

```python
plt.figure(figsize=(14, 6))
plt.boxplot([data[data['DriveTrain'] == drivetrain]['MPG_City'] for drivetrain in data['DriveTrain'].unique()],
            labels=data['DriveTrain'].unique(), vert=False)
plt.title('Boxplot de MPG en Ciudad por Tipo de Traccin (DriveTrain)')
plt.xlabel('MPG en Ciudad')
plt.ylabel('Tipo de Traccin (DriveTrain)')
plt.show()
```

```python
matrixCor = pd.DataFrame(data[["MSRP","Invoice","EngineSize","Horsepower","Cylinders","MPG_Highway","MPG_City"]])
matrixCor.corr()
```

Las variables ms importantes para explicar la variabilidad en el rendimiento son el tamao del motor, los caballos de fuerza, la cantidad de cilindros y las millas en autopista. Cuando una variable es positiva, significa que cuando una de las dos variables aumenta, la otra tambin aumenta. En cambio, una relacin negativa indica que si una variable aumenta, la otra disminuye. Esto se observa en el scatterplot, donde se nota que el tamao del motor est relacionado con un menor rendimiento de gasolina en la ciudad, es decir, a mayor tamao de motor, menor eficiencia en el consumo de gasolina en reas urbanas.
