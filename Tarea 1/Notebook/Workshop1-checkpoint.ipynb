```python
def shapiroTest(data):
    stat, p = stats.shapiro(data)
    alpha = 0.05

    if p > alpha:
        print("Los datos parecen seguir una distribución normal (no se rechaza la hipótesis nula)")
    else:
        print("Los datos no siguen una distribución normal (se rechaza la hipótesis nula)")
```

```python
shapiroTest(data['MSRP'])
shapiroTest(data['Invoice'])
shapiroTest(data['EngineSize'])
shapiroTest(data['Cylinders'])
shapiroTest(data['Horsepower'])
shapiroTest(data['MPG_City'])
shapiroTest(data['MPG_Highway'])
shapiroTest(data['Weight'])
shapiroTest(data['Wheelbase'])
shapiroTest(data['Length'])
```

```python
plt.figure(figsize=(12, 6))
plt.boxplot([data[data['Type'] == t]['MPG_City'] for t in data['Type'].unique()],
            labels=data['Type'].unique(), vert=False)
plt.title('Boxplot de MPG en Ciudad por Tipo de Vehículo')
plt.xlabel('MPG en Ciudad')
plt.ylabel('Tipo de Vehículo')
plt.show()
```

El gráfico de caja muestra que la mayoría de los datos están cerca del medio en la mayoría de los autos, pero en los Sedanes hay algunos valores extremos que están muy lejos de la mayoría, incluso fuera del rango normal. En cambio, en los autos híbridos, la mayoría de los datos están en el extremo más bajo.

```python
plt.figure(figsize=(14, 6))
plt.boxplot([data[data['Make'] == make]['MPG_City'] for make in data['Make'].unique()],
            labels=data['Make'].unique(), vert=False)
plt.title('Boxplot de MPG en Ciudad por Fabricante (Make)')
plt.xlabel('MPG en Ciudad')
plt.ylabel('Fabricante (Make)')
plt.show()
```

Las gráficas varían mucho entre las diferentes marcas de automóviles. En algunos casos, como Honda, hay solo un valor atípico, pero en otros, como Kia, la mayoría de los datos se encuentran en el extremo inferior. Los gráficos de caja muestran que los valores de las millas por galón (MPG_City) son similares en la mayoría de las marcas, pero algunas marcas tienen valores muy diferentes en este aspecto.

```python
plt.figure(figsize=(14, 6))
plt.boxplot([data[data['Origin'] == origin]['MPG_City'] for origin in data['Origin'].unique()],
            labels=data['Origin'].unique(), vert=False)
plt.title('Boxplot de MPG en Ciudad por Origen (Origin)')
plt.xlabel('MPG en Ciudad')
plt.ylabel('Origen (Origin)')
plt.show()
```

Los valores promedio de MPG_City son muy similares para los diferentes lugares de origen, lo que indica que la eficiencia de las millas por galón de los vehículos no está relacionada con su lugar de construcción. Sin embargo, en todos los casos, se observan varios valores atípicos que son excepciones a esta tendencia.

```python
plt.figure(figsize=(14, 6))
plt.boxplot([data[data['DriveTrain'] == drivetrain]['MPG_City'] for drivetrain in data['DriveTrain'].unique()],
            labels=data['DriveTrain'].unique(), vert=False)
plt.title('Boxplot de MPG en Ciudad por Tipo de Tracción (DriveTrain)')
plt.xlabel('MPG en Ciudad')
plt.ylabel('Tipo de Tracción (DriveTrain)')
plt.show()
```

En el gráfico de caja para "Front" se observan varios valores atípicos, lo que indica que hay datos poco confiables en esta columna. Lo mismo ocurre en "Rear", donde hay outliers tanto en la parte inferior como en la superior, lo que sugiere que algunos datos no son fiables. Por otro lado, en "All" no se ven outliers, lo que significa que los datos en esta categoría son más confiables para su comparación.

```python
matrixCor = pd.DataFrame(data[["MSRP","Invoice","EngineSize","Horsepower","Cylinders","MPG_Highway","MPG_City"]])
matrixCor.corr()
```

Las variables más importantes para explicar la variabilidad en el rendimiento son el tamaño del motor, los caballos de fuerza, la cantidad de cilindros y las millas en autopista. Cuando una variable es positiva, significa que cuando una de las dos variables aumenta, la otra también aumenta. En cambio, una relación negativa indica que si una variable aumenta, la otra disminuye. Esto se observa en el scatterplot, donde se nota que el tamaño del motor está relacionado con un menor rendimiento de gasolina en la ciudad, es decir, a mayor tamaño de motor, menor eficiencia en el consumo de gasolina en áreas urbanas.
